```python
# ç²å–è³‡æ–™
import requests
import json
r1 = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
pop_data = json.loads(r1.text)
r2 = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
nonpop_data = json.loads(r2.text)
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
len(pop_data) # ç†±é–€æ–‡ç« ä¸€æ¬¡å›å‚³30ç­†è³‡æ–™
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
# æ ¹æ“šä¸Šé¢çš„è‹±æ–‡åç¨±æ¬„ä½æ¨æ¸¬
for i in pop_data:
    print('æ¨™é¡Œ:{} è²¼æ–‡æ™‚é–“:{} ç•™è¨€äººæ•¸:{} æŒ‰è®šäººæ•¸:{}'.format(i['title'], i['createdAt'], \
                                                 i['commentCount'], i['likeCount']))
```

```python
# è¨ˆç®—å¹³å‡
def average(data, index):
    l = []
    for i in data:
        l.append(i[index])
    return sum(l)/len(data)
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
print('ç†±é–€æ–‡ç« -å¹³å‡ç•™è¨€äººæ•¸:{}ï¼Œå¹³å‡æŒ‰è®šäººæ•¸:{}'.format(average(pop_data, 'commentCount'), \
                                       average(pop_data, 'likeCount')))
print('éç†±é–€æ–‡ç« -å¹³å‡ç•™è¨€äººæ•¸:{}ï¼Œå¹³å‡æŒ‰è®šäººæ•¸:{}'.format(average(nonpop_data, 'commentCount'), \
                                       average(nonpop_data, 'likeCount')))
```

### 爬取 [Dcard熱門文章標題](https://www.dcard.tw/f)

## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'
response = r.text
# 測試是否取得資料
response[:600]
```

```python
soup = BeautifulSoup(response, 'html5lib')
soup
```

```python
for i in soup.find_all('h3', class_ = 'Title__Text-v196i6-0 gmfDU'):
    print(i.text)
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq
```

```python
g = Grab()
resp = g.go(url)
resp.body
```

```python
doc = pq(resp.body)
article_title = doc('h3')
article_title
# 根據h3的class，正好就是要取得的標題名稱
```

```python
for s in article_title:
    print(s.text)
```

### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text, d.find_all('em')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# dict
article = {}
# {類別: [文章標題]}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find_all('em')[-1].text in article.keys(): # 已經出現過的文章類別
        article[d.find_all('em')[-1].text].append(d.find_all('a')[-1].text)
    else: # 第一次出現的文章類別
        article[d.find_all('em')[-1].text] = [d.find_all('a')[-1].text]
               
article
```

### ③ 哪一個類別的文章最多

```python
# 每個類別的文章數目
count = {key : len(value) for key,value in article.items()}
count
```

```python
max(count.items())
```
